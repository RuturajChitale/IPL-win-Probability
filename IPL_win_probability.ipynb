```python
total_score_df =matches.groupby(['match_id', 'innings'])['runs_total'].sum().reset_index()
```

```python
total_score_df = total_score_df[total_score_df['innings'] == 2]
```

```python
match_df = matches.merge(total_score_df[['match_id','runs_total']],left_on='match_id', right_on = 'match_id')
```

```python
match_df['batting_team'].unique()
```

```python
import numpy as np

teams = np.array([
    "Chennai Super Kings",
    "Mumbai Indians",
    "Kolkata Knight Riders",
    "Royal Challengers Bengaluru",
    "Sunrisers Hyderabad",
    "Rajasthan Royals",
    "Delhi Capitals",
    "Punjab Kings",
    "Lucknow Super Giants",
    "Gujarat Titans"
])
```

```python
match_df['batting_team'] = match_df['batting_team'].replace({
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
})

match_df['bowling_team'] = match_df['bowling_team'].replace({
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
})
```

```python
valid_matches = match_df[
    match_df['batting_team'].isin(teams) &
    match_df['bowling_team'].isin(teams)
]['match_id']

match_df = match_df[match_df['match_id'].isin(valid_matches)]
```

```python
match_df.shape
```

```python
match_df['season']
```

```python

match_df['runs_left'] = match_df['runs_target'] - match_df['team_runs']

match_df['balls_left'] = 120 - ((match_df['over'] - 1) * 6 + match_df['ball'])

match_df['wickets_left'] = 10 - match_df['team_wicket']

match_df['total_runs_x'] = match_df['runs_target'] - 1

balls_played = 120 - match_df['balls_left']
match_df['crr'] = (match_df['team_runs'] * 6) / balls_played.replace(0, 1)

match_df['rrr'] = (match_df['runs_left'] * 6) / match_df['balls_left'].replace(0, 1)

match_df['result'] = (
    match_df['batting_team'] == match_df['match_won_by']
).astype(int)
```

```python
match_df.shape
```

```python
match_df['innings'].unique
```

```python
match_df = match_df[match_df['innings']==2]
```

```python
match_df.shape
```

```python
match_df['team_wicket']
```

```python
 match_df['result']
```

```python
sample_df = match_df[[
    'batting_team',
    'bowling_team',
    'city',
    'runs_left',
    'balls_left',
    'wickets_left',
    'total_runs_x',
    'crr',
    'rrr',
    'result'
]]
```

```python
sample_df.sample(sample_df.shape[0])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = sample_df.iloc[:, :-1]
Y = sample_df.iloc[:, -1]
X_train, X_test,Y_train, Y_test = train_test_split( X,Y,test_size=0.2, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder(sparse_output=False, drop='first'),
         ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1' , trf),
    ('step2' ,LogisticRegression(solver = 'liblinear'))
     ])
```

```python
pipe.fit(X_train,Y_train)
```

```python
X_train.describe()
```

```python
Y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```
